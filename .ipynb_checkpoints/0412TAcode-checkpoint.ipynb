```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
```

```python
x_data = boston.data
y_data = boston.target
```

```python
x_data[0]
```

```python
y_data[0]
```

```python
boston.feature_names
```

```python
#分離資料
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state = 123)
#開線性回歸
regr = LinearRegression()
#訓練
regr.fit(x_train, y_train)
```

```python
regr.intercept_
```

```python
#係數
regr.coef_
```

## SVM

```python
from sklearn.svm import SVC
```

```python
x = np.array([[-3, 1], [-6, 7], [3, -2], [5, -8]])
```

```python
y = np.array([1, 1, 2, 2]) #設定每筆資料對應的資料種類是第幾種→資料分類
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y)
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict([[2.5, 3]]) #輸入讓電腦去判斷他是第幾類 他覺得是第2類
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
X[0]
```

```python
Y = iris.target
```

```python
Y[0]
```

```python
Y #全部的種類，種共有3種(0, 1, 2)
```

```python
x_data = X[:, 1:3] #全取,中間兩個值
x_data[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, Y, test_size = 0.2, random_state = 111)
```

```python
plt.scatter(x_train[:, 0],x_train[:, 1], c = y_train) #x的前面的值放在x軸，x的後面的值放在y軸
```

```python
clf = SVC()
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
plt.scatter(x_test[:, 0],x_test[:, 1], c = y_test)
```

### 看預測的準不準

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predict) #對角線是正確的，預測是分類是1，答案是1有6個
```
