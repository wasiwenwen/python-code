# 利用Gradient Descent找出函數的最小值

## 讀進基本套件

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random as rd
```

## 決定Loss function

#### $f(x) = (x - 9)(x - 4)(x + 8)(x + 7)$

```python
def LossFunction(x):
    return  (x - 9) * (x - 4) * (x + 8) * (x + 7)

x = np.linspace(-10, 10, 100)
y = LossFunction(x)

plt.plot(x, y)
```

## 對Loss function做微分
#### $f'(x) = 4x^3 + 6x^2 -206x - 188$

```python
def dLossfunction(x):
    return 4 * x ** 3 + 6  * x ** 2 - 206 * x - 188
```

## Gradient Descent

```python
def GradientDescent(w, LearningRate):
    w -= dLossfunction(w) * LearningRate
    return w

start = round(rd.uniform(-10, 10), 4) #隨機選取初始值，到小數點後四位
print(start)
```

```python
LR = 0.0005 #learning rate，先隨便選一個數字
N = 100 #最多做一百次

wlist = [start]
ylist = [LossFunction(start)] 
clist = [0] #根據次數改變顏色，從紫到黃

point = start
for i in range(0, N):
    point = round(GradientDescent(point, LR), 4) #取到小數點後四位
    if point in wlist: break #重複了就停止
    wlist.append(point)
    ylist.append(LossFunction(point))
    clist.append(i // 5)
    
plt.plot(x, y)
plt.scatter(wlist, ylist, c = clist)
plt.scatter(wlist[-1], ylist[-1], c = 'r') #標出最後的點，為紅色
print('最後走到的區域最小值為%.4f，此時w為%.4f' %(ylist[-1], wlist[-1]))
print('過程中有找到的區域最小值為%.4f' %min(ylist))
```

### 若是稍微改變一下Learning rate呢？
#### 稍微設大一點→可能會不停錯過最低點，亂跳，沒有好好收斂，最後不是停在最低點

```python
LR = 0.006 #learning rate
N = 100 #最多做一百次

wlist1 = [start]
ylist1 = [LossFunction(start)] 
clist1 = [0] #根據次數改變顏色，從紫到黃

point = start
for i in range(0, N):
    point = round(GradientDescent(point, LR), 4) #取到小數點後四位
    if point in wlist1: break #重複了就停止
    wlist1.append(point)
    ylist1.append(LossFunction(point))
    clist1.append(i // 5)

plt.plot(x, y)
plt.scatter(wlist1, ylist1, c = clist1)
plt.scatter(wlist1[-1], ylist1[-1], c = 'r') #標出最後的點，為紅色
print('最後走到的區域最小值為%.4f，此時w為%.4f' %(ylist1[-1], wlist1[-1]))
print('過程中有找到的區域最小值為%.4f' %min(ylist1))
```

#### 稍微設小一點→可能會移動緩慢花了100次還到不了最低點

```python
LR = 0.0001 #learning rate
N = 100 #最多做一百次

wlist2 = [start]
ylist2 = [LossFunction(start)] 
clist2 = [0] #根據次數改變顏色，從紫到黃

point = start
for i in range(0, N):
    point = round(GradientDescent(point, LR), 4) #取到小數點後四位
    if point in wlist2: break #重複了就停止
    wlist2.append(point)
    ylist2.append(LossFunction(point))
    clist2.append(i // 5)
    
plt.plot(x, y)
plt.scatter(wlist2, ylist2, c = clist2)
plt.scatter(wlist2[-1], ylist2[-1], c = 'r') #標出最後的點，為紅色
print('最後走到的區域最小值為%.4f，此時w為%.4f' %(ylist2[-1], wlist2[-1]))
print('過程中有找到的區域最小值為%.4f' %min(ylist2))
```

#### 結論與發現：設定適當的learning rate，可以讓人事半功倍，正確的一步步走向區域最小值，太大或太小都有可能錯過最小值
